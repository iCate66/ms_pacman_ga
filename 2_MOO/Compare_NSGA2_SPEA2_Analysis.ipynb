# Multi-Objective Optimisation Analysis: NSGA-II vs SPEA2 for Ms. Pac-Man

This notebook provides a comparative analysis of NSGA-II and SPEA2 algorithms applied to Ms. Pac-Man optimisation. Various metrics will be analysed including game score, complexity, convergence rate, solution diversity, and Pareto fronts.

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.spatial.distance import pdist, squareform
import pickle
import warnings
warnings.filterwarnings('ignore')

# Load results from pickle files
def load_results(nsga2_path, spea2_path):
    """Load results from both algorithms"""
    with open(nsga2_path, 'rb') as f:
        nsga2_results = pickle.load(f)
    with open(spea2_path, 'rb') as f:
        spea2_results = pickle.load(f)
    return nsga2_results, spea2_results

# Load the results
# nsga2_results, spea2_results = load_results('/Users/catherineiwata/Desktop/Final_Project/code_folder/multi_processing/runs_results_MOO_TEST/2_NSGA2_DERIVED_RESULTS/run_1_5110/pacman_nsga2_final_results.pkl', 
#                                           '/Users/catherineiwata/Desktop/Final_Project/code_folder/multi_processing/runs_results_MOO_TEST/3_SPEA2_DERIVED_results/run_1/pacman_spea2_final_results.pkl')
nsga2_results, spea2_results = load_results('NSGA2/run_1_5110/pacman_nsga2_final_results.pkl', 
                                            'SPEA2/run_1_4590/pacman_spea2_final_results.pkl')
```

```python
# Helper functions for metrics calculation

def calculate_hypervolume(points, ref_point):
    """Calculate hypervolume indicator"""
    points = np.array(points)
    points = points[np.all(points <= ref_point, axis=1)]
    if len(points) == 0:
        return 0
    points = np.sort(points, axis=0)
    volume = 0
    for i in range(len(points)):
        if i == 0:
            v = np.prod(ref_point - points[i])
        else:
            v = np.prod(np.minimum(ref_point, np.maximum(points[i], points[i-1])) - points[i-1])
        volume += v
    return volume

def calculate_spread(points):
    """Calculate spread metric (diversity)"""
    if len(points) < 2:
        return 0
    distances = pdist(points)
    return np.std(distances) / np.mean(distances)
```

```python
# Analyse and compare the game scores achieved by both algorithms

# 1. Game Score Analysis
def analyse_game_scores(nsga2_results, spea2_results):
    """Analyse and compare game scores between algorithms"""
    
    # Calculate mean and max scores across generations
    nsga2_mean_scores = []
    nsga2_max_scores = []
    spea2_mean_scores = []
    spea2_max_scores = []
    
    for gen in nsga2_results['generation_history']:
        scores = [m['total_score'] for m in gen['metrics'] if m]
        if scores:
            nsga2_mean_scores.append(np.mean(scores))
            nsga2_max_scores.append(np.max(scores))
            
    for gen in spea2_results['generation_history']:
        scores = [m['total_score'] for m in gen['metrics'] if m]
        if scores:
            spea2_mean_scores.append(np.mean(scores))
            spea2_max_scores.append(np.max(scores))
    
    # Create subplots
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
    
    # 1. Final Generation Score Distribution (Violin Plot)
    nsga2_final_scores = [m['total_score'] for m in nsga2_results['generation_history'][-1]['metrics']]
    spea2_final_scores = [m['total_score'] for m in spea2_results['generation_history'][-1]['metrics']]
    
    final_scores_df = pd.DataFrame({
        'Algorithm': ['NSGA-II'] * len(nsga2_final_scores) + ['SPEA2'] * len(spea2_final_scores),
        'Score': nsga2_final_scores + spea2_final_scores
    })
    
    sns.violinplot(data=final_scores_df, x='Algorithm', y='Score', ax=ax1)
    ax1.set_title('Final Generation Score Distribution')
    ax1.set_ylabel('Game Score')
    
    # 2. Mean Scores Over Generations
    generations = range(len(nsga2_mean_scores))
    ax2.plot(generations, nsga2_mean_scores, label='NSGA-II', alpha=0.8)
    ax2.plot(generations, spea2_mean_scores, label='SPEA2', alpha=0.8)
    ax2.set_title('Mean Scores Over Generations')
    ax2.set_xlabel('Generation')
    ax2.set_ylabel('Mean Score')
    ax2.legend()
    ax2.grid(True)
    
    # 3. Max Scores Over Generations
    ax3.plot(generations, nsga2_max_scores, label='NSGA-II', alpha=0.8)
    ax3.plot(generations, spea2_max_scores, label='SPEA2', alpha=0.8)
    ax3.set_title('Maximum Scores Over Generations')
    ax3.set_xlabel('Generation')
    ax3.set_ylabel('Max Score')
    ax3.legend()
    ax3.grid(True)
    
    # 4. Box plot of final generation scores
    sns.boxplot(data=final_scores_df, x='Algorithm', y='Score', ax=ax4)
    ax4.set_title('Final Generation Score Distribution (Box Plot)')
    ax4.set_ylabel('Game Score')
    
    plt.tight_layout()
    
    # Statistical tests
    t_stat, p_value = stats.ttest_ind(nsga2_final_scores, spea2_final_scores)
    
    print("\nScore Statistics:")
    print(f"Final Generation Metrics:")
    print(f"NSGA-II - Mean: {np.mean(nsga2_final_scores):.2f}, Std: {np.std(nsga2_final_scores):.2f}")
    print(f"        Max: {np.max(nsga2_final_scores):.2f}, Min: {np.min(nsga2_final_scores):.2f}")
    print(f"SPEA2   - Mean: {np.mean(spea2_final_scores):.2f}, Std: {np.std(spea2_final_scores):.2f}")
    print(f"        Max: {np.max(spea2_final_scores):.2f}, Min: {np.min(spea2_final_scores):.2f}")
    print(f"\nT-test results - t-statistic: {t_stat:.3f}, p-value: {p_value:.3f}")
    
    # Calculate and print overall best scores
    print("\nOverall Best Scores:")
    print(f"NSGA-II - Best ever: {max(nsga2_max_scores):.2f}")
    print(f"SPEA2   - Best ever: {max(spea2_max_scores):.2f}")
    
    plt.show()
    return t_stat, p_value, {
        'nsga2_mean_scores': nsga2_mean_scores,
        'nsga2_max_scores': nsga2_max_scores,
        'spea2_mean_scores': spea2_mean_scores,
        'spea2_max_scores': spea2_max_scores
    }

# Run game score analysis
score_stats = analyse_game_scores(nsga2_results, spea2_results)
```

```python
# 2. Complexity Analysis

# Analyse the complexity of solutions produced by each algorithm.
def analyse_complexity(nsga2_results, spea2_results):
    """Analyse solution complexity based on strategy variations"""
    def get_complexity_metrics(results, algorithm_name):
        """Extract complexity metrics"""
        complexities = []
        
        # Debug print
        print(f"\nAnalysing {algorithm_name} data structure:")
        print(f"Keys in results: {results.keys()}")
        print(f"Length of generation history: {len(results['generation_history'])}")
        
        # Print complete metrics structure
        print("\nMetrics structure:")
        example_metrics = results['generation_history'][-1]['metrics'][0]
        print(f"Available metrics keys: {example_metrics.keys()}")
        
        for i, metrics in enumerate(results['generation_history'][-1]['metrics']):
            if not metrics:
                continue
                
            # Calculate complexity based on scoring patterns and strategy changes
            complexity_score = 0
            
            # 1. Score pattern complexity
            if 'score_segments' in metrics:
                score_changes = np.diff([0] + metrics['score_segments'])
                significant_changes = [x for x in score_changes if abs(x) > 50]
                pattern_complexity = len(significant_changes) / max(1, len(metrics['score_segments']))
                complexity_score += pattern_complexity * 3  # Weight factor
            
            # 2. Ghost hunting complexity
            if 'ghost_points' in metrics and 'pellet_points' in metrics:
                total_points = metrics['ghost_points'] + metrics['pellet_points']
                if total_points > 0:
                    ghost_ratio = metrics['ghost_points'] / total_points
                    # Higher complexity for balanced ghost/pellet strategies
                    balance_factor = 4 * (ghost_ratio * (1 - ghost_ratio))  # Peaks at 0.5 ratio
                    complexity_score += balance_factor
            
            # 3. Survival strategy complexity
            if 'deaths' in metrics and 'frames_survived' in metrics:
                survival_ratio = metrics['frames_survived'] / (max(1, metrics['deaths'] + 1))
                survival_complexity = min(3, survival_ratio / 1000)  # Cap at 3
                complexity_score += survival_complexity
            
            complexities.append(complexity_score)
            
            # Debug first few entries
            if i < 3:
                print(f"\nExample {i+1} complexity breakdown:")
                print(f"Total complexity score: {complexity_score:.2f}")
                if 'score_segments' in metrics:
                    print(f"Score segments length: {len(metrics['score_segments'])}")
                if 'ghost_points' in metrics and 'pellet_points' in metrics:
                    print(f"Ghost points: {metrics['ghost_points']}")
                    print(f"Pellet points: {metrics['pellet_points']}")
                if 'deaths' in metrics:
                    print(f"Deaths: {metrics['deaths']}")
                if 'frames_survived' in metrics:
                    print(f"Frames survived: {metrics['frames_survived']}")
        
        print(f"\n{algorithm_name} complexity stats:")
        print(f"Number of valid complexity scores: {len(complexities)}")
        if complexities:
            print(f"Range: {min(complexities):.2f} - {max(complexities):.2f}")
            print(f"Mean: {np.mean(complexities):.2f}")
            print(f"Std: {np.std(complexities):.2f}")
        
        return complexities
    
    # Get complexity metrics with debug info
    nsga2_complexity = get_complexity_metrics(nsga2_results, "NSGA-II")
    spea2_complexity = get_complexity_metrics(spea2_results, "SPEA2")
    
    # Check valid data
    if not nsga2_complexity or not spea2_complexity:
        print("Error: No valid complexity metrics found for one or both algorithms")
        return None, None
    
    # Create DataFrame 
    complexity_df = pd.DataFrame({
        'Algorithm': ['NSGA-II'] * len(nsga2_complexity) + ['SPEA2'] * len(spea2_complexity),
        'Complexity': nsga2_complexity + spea2_complexity
    })
    
    # Create more visualisations
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Boxplot
    sns.boxplot(data=complexity_df, x='Algorithm', y='Complexity', ax=ax1)
    ax1.set_title('Complexity Distribution')
    ax1.set_ylabel('Complexity Score')
    
    # Violin plot with individual points
    sns.violinplot(data=complexity_df, x='Algorithm', y='Complexity', ax=ax2)
    sns.stripplot(data=complexity_df, x='Algorithm', y='Complexity', 
                 color='red', alpha=0.3, size=4, ax=ax2)
    ax2.set_title('Complexity Distribution with Individual Points')
    ax2.set_ylabel('Complexity Score')
    
    plt.tight_layout()
    
    # Statistical test
    t_stat, p_value = stats.ttest_ind(nsga2_complexity, spea2_complexity)
    print("\nStatistical Comparison:")
    print(f"NSGA-II - Mean: {np.mean(nsga2_complexity):.2f}, Std: {np.std(nsga2_complexity):.2f}")
    print(f"SPEA2 - Mean: {np.mean(spea2_complexity):.2f}, Std: {np.std(spea2_complexity):.2f}")
    print(f"T-test results - t-statistic: {t_stat:.3f}, p-value: {p_value:.3f}")
    
    # Effect size calculation
    cohens_d = (np.mean(nsga2_complexity) - np.mean(spea2_complexity)) / \
               np.sqrt((np.std(nsga2_complexity)**2 + np.std(spea2_complexity)**2) / 2)
    print(f"Cohen's d effect size: {cohens_d:.3f}")
    
    plt.show()
    return t_stat, p_value

# Run complexity analysis
complexity_stats = analyse_complexity(nsga2_results, spea2_results)
```

```python
# 3. Convergence Analysis

# Analyse speed each algorithm converges to solutions
def analyse_convergence(nsga2_results, spea2_results):
    """Analyse convergence rates"""
    def calculate_convergence_rate(history):
        """Calculate convergence rate based on hypervolume progress"""
        hypervolumes = []
        ref_point = np.array([1.1, 1.1, 1.1])  # Reference point for hypervolume
        for gen_data in history:
            points = np.array(gen_data['objective_values'])
            hv = calculate_hypervolume(points, ref_point)
            hypervolumes.append(hv)
        
        # Calculate rate as slope of hypervolume improvement
        generations = range(len(hypervolumes))
        slope, _, _, _, _ = stats.linregress(generations, hypervolumes)
        return slope, hypervolumes
    
    nsga2_rate, nsga2_hv = calculate_convergence_rate(nsga2_results['generation_history'])
    spea2_rate, spea2_hv = calculate_convergence_rate(spea2_results['generation_history'])
    
    plt.figure(figsize=(10, 6))
    plt.plot(nsga2_hv, label='NSGA-II')
    plt.plot(spea2_hv, label='SPEA2')
    plt.xlabel('Generation')
    plt.ylabel('Hypervolume')
    plt.title('Convergence Comparison')
    plt.legend()
    
    print("\nConvergence Statistics:")
    print(f"NSGA-II convergence rate: {nsga2_rate:.6f}")
    print(f"SPEA2 convergence rate: {spea2_rate:.6f}")
    
    plt.show()
    return nsga2_rate, spea2_rate

# Run convergence analysis
convergence_stats = analyse_convergence(nsga2_results, spea2_results)
```

```python
# 4. Diversity Analysis

# Analyse how well each algorithm maintains solution diversity
def analyse_diversity(nsga2_results, spea2_results):
    """Analyse solution diversity"""
    def calculate_generation_diversity(gen_data):
        points = np.array(gen_data['objective_values'])
        return calculate_spread(points)
    
    nsga2_diversity = [calculate_generation_diversity(gen) for gen in nsga2_results['generation_history']]
    spea2_diversity = [calculate_generation_diversity(gen) for gen in spea2_results['generation_history']]
    
    plt.figure(figsize=(10, 6))
    plt.plot(nsga2_diversity, label='NSGA-II')
    plt.plot(spea2_diversity, label='SPEA2')
    plt.xlabel('Generation')
    plt.ylabel('Diversity Metric')
    plt.title('Solution Diversity Over Generations')
    plt.legend()
    
    print("\nDiversity Statistics:")
    print(f"NSGA-II final diversity: {nsga2_diversity[-1]:.4f}")
    print(f"SPEA2 final diversity: {spea2_diversity[-1]:.4f}")
    
    plt.show()
    return nsga2_diversity[-1], spea2_diversity[-1]

# Run diversity analysis
diversity_stats = analyse_diversity(nsga2_results, spea2_results)
```

```python
# 5. Pareto Front Visualisation

# Visualise and compare the Pareto fronts from both algorithms
def visualise_pareto_fronts(nsga2_results, spea2_results):
    """Visualise Pareto fronts"""
    def get_pareto_front(results):
        points = np.array(results['generation_history'][-1]['objective_values'])
        return points[np.argsort(points[:, 0])]
    
    nsga2_front = get_pareto_front(nsga2_results)
    spea2_front = get_pareto_front(spea2_results)
    
    # 2D visualisation (score vs efficiency)
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.scatter(nsga2_front[:, 0], nsga2_front[:, 2], label='NSGA-II', alpha=0.6)
    plt.scatter(spea2_front[:, 0], spea2_front[:, 2], label='SPEA2', alpha=0.6)
    plt.xlabel('Score Objective')
    plt.ylabel('Efficiency Objective')
    plt.title('Pareto Front: Score vs Efficiency')
    plt.legend()
    
    # 3D visualisation
    plt.subplot(1, 2, 2, projection='3d')
    ax = plt.gca()
    ax.scatter(nsga2_front[:, 0], nsga2_front[:, 1], nsga2_front[:, 2], 
              label='NSGA-II', alpha=0.6)
    ax.scatter(spea2_front[:, 0], spea2_front[:, 1], spea2_front[:, 2], 
              label='SPEA2', alpha=0.6)
    
    ax.set_xlabel('Score')
    ax.set_ylabel('Survival')
    ax.set_zlabel('Efficiency')
    ax.set_title('3D Pareto Front')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    return nsga2_front, spea2_front

# Visualise Pareto fronts
pareto_fronts = visualise_pareto_fronts(nsga2_results, spea2_results)
```

```python
# 6. Summary Statistics of all metrics for easy comparison

# Create summary DataFrame
summary = pd.DataFrame({
    'Metric': ['Game Score', 'Complexity', 'Convergence Rate', 'Diversity'],
    'NSGA-II': [np.mean([m['total_score'] for m in nsga2_results['generation_history'][-1]['metrics']]),
                complexity_stats[0],
                convergence_stats[0],
                diversity_stats[0]],
    'SPEA2': [np.mean([m['total_score'] for m in spea2_results['generation_history'][-1]['metrics']]),
              complexity_stats[1],
              convergence_stats[1],
              diversity_stats[1]]
})

print("Summary Statistics:")
print(summary.to_string(index=False))
```
